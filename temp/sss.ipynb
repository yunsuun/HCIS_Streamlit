```python
import pandas as pd
```

```python

app_train_full_type_6 = pd.read_parquet(r"app_train_full_type_6.parquet")
oof_predictions_top70 = pd.read_parquet(r"oof_predictions_top70.parquet")
print('전체 파일 로드 완료')
```

```python
# 1. oof_predictions_top70에서 중복되는 'target' 컬럼 제거
oof_predictions_top70 = oof_predictions_top70.drop(columns=['target'], errors='ignore')

# 2. 병합 실행
df = pd.merge(app_train_full_type_6, oof_predictions_top70, on='sk_id_curr', how='left')

# 3. 컬럼 이름 변경
df = df.rename(columns={"p_hat_top70" : "pd_hat"})

# 확인
# 'target_x' 대신 'target'만 남고, 'target_y'는 없어집니다.
# 'app_train_full_type_7'의 target 컬럼이 'target'으로 유지됩니다.

df = pd.merge(app_train_full_type_6, oof_predictions_top70, on='sk_id_curr', how='left')
df = df.rename(columns={"p_hat_top70" : "pd_hat"})
print("pd_hat 컬럼명 변경 완료!")
```

```python
# inplace 옵션을 빼면 수정된 데이터프레임이 반환됩니다.
oof_predictions_top70 = oof_predictions_top70.rename(columns={"p_hat_top70": "pd_hat"})

# 확인
# oof_predictions_top70.info()
oof_predictions_top70.to_parquet("oof_predictions_top70.parquet")
print()
print('top 70 저장 완료')
```
